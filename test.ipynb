```python
from openai import OpenAI


VLLM_ENDPOINT = "202.92.159.240"   # Your server IP
VLLM_PORT = "8000"            # Your port number
MODEL_NAME = "Qwen/Qwen3.6-27B" # The model loaded in vLLM

# vLLM doesn't require a real API key by default, but the client 
# expects a string, so "EMPTY" or "none" is standard.
client = OpenAI(
    base_url=f"http://{VLLM_ENDPOINT}:{VLLM_PORT}/v1",
    api_key="EMPTY" 
)

response = client.chat.completions.create(
    model=MODEL_NAME,
    messages=[
         {"role": "user", "content": "Explain quantum entanglement in one sentence."}
    ],
    temperature=0.7,
    max_tokens=1000,
    extra_body={"chat_template_kwargs": {"enable_thinking": False}},
)


print(response.choices[0].message.content)
```
